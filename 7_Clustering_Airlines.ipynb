#### Perform clustering (hierarchical,K means clustering and DBSCAN) for the airlines data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering

import warnings 
warnings.filterwarnings('ignore')
```

# Hierarchical Clustering

```python
df = pd.read_csv("C:\\Users\\krush\\Desktop\\Data Science\\Assignments\\Clustering\\EastWestAirlines.csv", encoding='latin-1')
df
```

## EDA and Visualization

```python
df.rename(columns={'Award?':'Award', 'ID#':'ID' }, inplace= True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

## Scaling down the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = pd.DataFrame(sc.fit_transform(df.iloc[:,1: ]), columns=df.iloc[:,1: ].columns)
x
```

## Finding optimal number of clusters using dendograms

```python
from scipy.cluster import hierarchy
```

```python
lk = hierarchy.linkage(x, method='complete')
```

```python
dendrogram = hierarchy.dendrogram(lk)
plt.figure(figsize = (150,50))
plt.title('Hierarchical Dendrogram')
plt.show()
```

## Model Building

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5)
```

```python
ylabel = hc.fit_predict(x)
ylabel
```

```python
df['Cluster'] = ylabel
df
```

```python
df.groupby('Cluster').agg('mean')
```

## Visualizing the clusters

```python
df[df['Cluster']==2] 
```

```python
sns.scatterplot(df['Balance'], df['Cluster'], hue = df['ID'])
plt.legend(bbox_to_anchor = (1,1))
```

# Kmeans Clustering

```python
df2 = pd.read_csv("C:\\Users\\krush\\Desktop\\Data Science\\Assignments\\Clustering\\EastWestAirlines.csv", encoding='latin-1')
df2
```

## EDA and Visualization

```python
df2.rename(columns={'Award?':'Award', 'ID#':'ID' }, inplace= True)
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2.duplicated().sum()
```

## Scaling down the data

```python
x2 = pd.DataFrame(sc.fit_transform(df2.iloc[:,1: ]), columns=df2.iloc[:,1: ].columns)
x2
```

## Basic Model

```python
from sklearn.cluster import KMeans
km = KMeans()
```

```python
cluster2 = km.fit_predict(x2)
cluster2
```

## Finding optimal number of clusters using Elbow Graph

```python
wcss2 = []

for i in range(1,20):
    km = KMeans(n_clusters=i)
    km.fit_predict(x2)
    wcss2.append(km.inertia_)
```

```python
wcss2
```

```python
plt.plot(range(1,20), wcss2)
plt.scatter(range(1, 20), wcss2,color = 'red')
plt.title('ELBOW METHOD')
plt.xlabel('Kvalues')
plt.ylabel('WCSS')
plt.show()
```

## Revised Model

```python
# k-value = 6
km = KMeans(n_clusters=6)
ylabel2 = km.fit_predict(x2)
ylabel2
```

```python
df2['Cluster2'] = ylabel2
df2
```

## Visualizing the clusters

```python
df2[df2['Cluster2']==4]
```

```python
sns.scatterplot(df2['Balance'], df2['Cluster2'], hue = df2['ID'])
plt.legend(bbox_to_anchor = (1,1))
```

## Model Building

### Iteration 1

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps = 1,min_samples = 5)
```

```python
db.fit(x3)
```

```python
db.labels_
```

```python
df3['Cluster3'] = db.labels_
df3
```

```python
df3.Cluster3.value_counts()
```

### Iteration 2

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps = 0.8,min_samples = 10)
```

```python
db.fit(x3)
```

```python
db.labels_
```

```python
df3['Cluster3'] = db.labels_
df3
```

```python
df3.Cluster3.value_counts()
```
